```r
#1
df<- read.csv("~/Desktop/siminar project/ticks/Data+Host+Seeking+2019_BU.csv", stringsAsFactors=FALSE)
df[,c(1,2,4:15)] <- NULL
df[,c(2:5)] <- NULL
View(df)

#t=30min

#The percentage of ticks which remained at start is 0.
#The number of ticks which remained in q1, q2, q3 and q4 are 31,7, 2 and 5 respectively.

#Ho1: The probabilities of ticks remaining in the 4 quadrants are the same, p1=p2=p3=p4=0.25.
#Ha1: At least one of them are different from others.
x=c(31,7,2,5)
chisq.test(x)

#Based on the p-value of chi-square test, we can reject the Ho. The probabilities of ticks remaining in 4 quadrants are significant different.

#Ho2: The probabilities of ticks remaining in the quadrant 1 is same as the sum of the other three, p1=p2+p3+p4=0.5.
#Ha2: The probabilities of ticks remaining in the quadrant 1 is much higher than the sum of the other three.

y=c(31,14)
chisq.test(y)

#Based on the p-value of chi-square test, we can reject the Ho. The probability of ticks remaining in 1 quadrants are significant larger than the other three.

```

```r
#2
df2<-read.csv("~/Desktop/siminar project/project/Data+Host+Seeking+2019_BU.csv", stringsAsFactors=FALSE)
df2[,c(1,2,4:10)]<-NULL
View(df2)

#The average times which ticks show up at the start point during 30 mins.
mean(df2$NUMBER.OF.OBSERVATIONS.ON.START..out.of.30.)

#The probabilities of the ticks which show up in the 4 quadrants
st=df2$NUMBER.OF.OBSERVATIONS.ON.START..out.of.30.
dim(st)=c(45,1)
q1=df2$NUMBER.OF.OBSERVATIONS.in.QUAD.1.out.of.30.
dim(q1)=c(45,1)
q2=df2$NUMBER.OF.OBSERVATIONS.in.QUAD.2..out.of.30.
dim(q2)=c(45,1)
q3=df2$NUMBER.OF.OBSERVATIONS.in.QUAD.3..out.of.30.
dim(q3)=c(45,1)
q4=df2$NUMBER.OF.OBSERVATIONS.in.QUAD.4.out.of.30.
dim(q4)=c(45,1)

X = cbind(q1,q2,q3,q4)
XX = X/rowSums(X)
plot(XX[,1],XX[,2])
plot(XX[,1],XX[,3])
plot(XX[,2],XX[,3])
colMeans(XX) - qnorm(0.9995)*apply(XX, 2, sd)/sqrt(45)
colMeans(XX) + qnorm(0.9995)*apply(XX, 2, sd)/sqrt(45)
```

```r
#3
#Comparing the maximum horizontal distance the tick traveled in quadrant 1 and not in quadrant 1.
#We use the Mann-Whitney test to check if the two sample have same distribution. 
#Ho:the distribution of the maximum horizontal distance the tick traveled in quadrant 1 and not in quadrant 1 are the same.
#Ha:the distribution is not the same.

dis_1=df2$MAXIMUM.HORIZONTAL.DISTANCE.TRAVELED.in.QUAD.1..cm....toward.host
dis_2=df2$MAXIMUM.HORIZONTAL.DISTANCE.TRAVELED.in.QUAD.2..cm.
dis_3=df2$MAXIMUM.HORIZONTAL.DISTANCE.TRAVELED.in.QUAD.3..cm....away.from.host
dis_4=df2$MAXIMUM.HORIZONTAL.DISTANCE.TRAVELED.in.QUAD.4.cm.
dim(dis_2)=c(45,1)
dim(dis_3)=c(45,1)
dim(dis_4)=c(45,1)
dis=cbind(dis_2,dis_3,dis_4)


dis_not1=vector()
for (i in 1:45) {
  dis_not1[i]=max(dis[i,])
}
dis_not1

#Mann-Whitney test
wilcox.test(dis_1,dis_not1)

#Based on the result above, the p-value here is 0.036, which is smaller than 0.05. We can reject the Ho, the distribution of the maximum horizontal distance the tick traveled in quadrant 1 and not in quadrant 1 are not same.
```

```r
#4
#horizontal

horizontal_dist <- df[,c(2:7)]
horizontal_dist <- as.matrix(horizontal_dist)
horizontal_dist <- t(horizontal_dist)
dim(horizontal_dist) <- c(270,1)

start <- seq(1,270, by=6)
diff_horiz <- vector()
any(start==i)
for(i in (1:270)){
  if(i%in%start){
    diff_horiz[i] <- horizontal_dist[i]
  }
  else{
    diff_horiz[i] <- horizontal_dist[i] - horizontal_dist[i-1]
  }
}

t <- gl(6,1,270)

ter <- df$TERRAIN.LEVEL
ter <- rep(ter,6)
dim(ter) <-c(45,6)
ter <- t(ter)
dim(ter) <- c(270,1)



# mean in plot below
dim(diff_horiz)=c(6,45)
(mean_1=colMeans(t(diff_horiz)[c(1:15),]))
(mean_2=colMeans(t(diff_horiz)[c(16:30),]))
(mean_3=colMeans(t(diff_horiz)[c(31:45),]))

# Plot the mean of the three terrain levels in the scattor plot
tt=seq(5,30,by=5)
plot(x=as.numeric(t),y=diff_horiz,xaxt="n",yaxt="n",xlab="time",ylab="incremental horizontal distance ",pch=16)
par(new=TRUE)
plot(tt,mean_1,type = "l",col=2,xlab="time",ylab="incremental horizontal distance ",ylim=c(-15,35))
par(new=TRUE)
plot(tt,mean_2,type = "l",col=3,xlab="time",ylab="incremental horizontal distance ",ylim=c(-15,35))
par(new=TRUE)
plot(tt,mean_3,type = "l",col=4,xlab="time",ylab="incremental horizontal distance ",ylim=c(-15,35))

legend("top",legend = c("Terrain 1","Terrain 2","Terrain 3"),ncol=3,cex=0.8,bty="n",col = c("red","green","blue"),lty=1,lwd=2)




mod1 <- lm(diff_horiz~as.factor(ter))
summary(mod1)


qqnorm(mod1$residuals)
qqline(mod1$residuals)

mod11 <- lm(diff_horiz~as.factor(ter)+t+as.factor(ter)*t)
summary(mod11)
par(mfrow=c(2,2))
plot(mod11)

#The scatter plot of the incremental horizontal distance - time for terrain 1
library(ggplot2)
t1=t[1:90]
diff1=diff_horiz[1:90]
data1=cbind(t1,diff1)
data1=data.frame(data1)
ggplot(data1,aes(x=t1,y=diff1))+geom_point()+geom_smooth(method=lm)+labs(title="terrain 1",x="time",y="incremental horizontal distance")+theme(plot.title = element_text(hjust = 0.5))


#The scatter plot of the incremental horizontal distance - time for terrain 2
t2=t[91:180]
diff2=diff_horiz[91:180]
data2=cbind(t2,diff2)
data2=data.frame(data2)
ggplot(data2,aes(x=t2,y=diff2))+geom_point()+geom_smooth(method = lm)+labs(title="terrain 2",x="time",y="incremental horizontal distance")+theme(plot.title = element_text(hjust = 0.5))


#The scatter plot of the incremental horizontal distance - time for terrain 3
t3=t[181:270]
diff3=diff_horiz[181:270]
data3=cbind(t3,diff3)
data3=data.frame(data3)
ggplot(data3,aes(x=t3,y=diff3))+geom_point()+geom_smooth(method = lm)+labs(title="terrain 3",x="time",y="incremental horizontal distance")+theme(plot.title = element_text(hjust = 0.5))



#vertical

vertical_dist <- df[,c(8:13)]
vertical_dist <- as.matrix(vertical_dist)
vertical_dist <- t(vertical_dist)
dim(vertical_dist) <- c(270,1)

start <- seq(1,270, by=6)
diff_verti <- vector()
any(start==i)
for(i in (1:270)){
  if(i%in%start){
    diff_verti[i] <- vertical_dist[i]
  }
  else{
    diff_verti[i] <- vertical_dist[i] - vertical_dist[i-1]
  }
}


mod2 <- lm(diff_verti~as.factor(ter)+t+as.factor(ter)*t)
summary(mod2)

qqnorm(mod2$residuals)
qqline(mod2$residuals)




# mean in plot below
dim(diff_verti)=c(6,45)
(mean_1=colMeans(t(diff_verti)[c(1:15),]))
(mean_2=colMeans(t(diff_verti)[c(16:30),]))
(mean_3=colMeans(t(diff_verti)[c(31:45),]))

# Plot the mean of the three terrain levels in the scattor plot
tt=seq(5,30,by=5)
plot(x=as.numeric(t),y=diff_verti,xaxt="n",yaxt="n",xlab="time",ylab="incremental vertical distance ",pch=16)
par(new=TRUE)
plot(tt,mean_1,type = "l",col=2,xlab="time",ylab="incremental vertical distance ",ylim=c(-10,15))
par(new=TRUE)
plot(tt,mean_2,type = "l",col=3,xlab="time",ylab="incremental vertical distance ",ylim=c(-10,15))
par(new=TRUE)
plot(tt,mean_3,type = "l",col=4,xlab="time",ylab="incremental vertical distance ",ylim=c(-10,15))

legend("top",legend = c("Terrain 1","Terrain 2","Terrain 3"),ncol=3,cex=0.8,bty="n",col = c("red","green","blue"),lty=1,lwd=2)




#The scatter plot of the incremental vertical distance - time for terrain 1
library(ggplot2)
t1=t[1:90]
diff1=diff_verti[1:90]
data1=cbind(t1,diff1)
data1=data.frame(data1)
ggplot(data1,aes(x=t1,y=diff1))+geom_point()+geom_smooth(method = lm)+labs(title="terrain 1",x="time",y="incremental vertical distance")+theme(plot.title = element_text(hjust = 0.5))


#The scatter plot of the incremental vertical distance - time for terrain 2
t2=t[91:180]
diff2=diff_verti[91:180]
data2=cbind(t2,diff2)
data2=data.frame(data2)
ggplot(data2,aes(x=t2,y=diff2))+geom_point()+geom_smooth(method = lm)+labs(title="terrain 2",x="time",y="incremental vertical distance")+theme(plot.title = element_text(hjust = 0.5))


#The scatter plot of the incremental vertical distance - time for terrain 3
t3=t[181:270]
diff3=diff_verti[181:270]
data3=cbind(t3,diff3)
data3=data.frame(data3)
ggplot(data3,aes(x=t3,y=diff3))+geom_point()+geom_smooth(method = lm)+labs(title="terrain 3",x="time",y="incremental vertical distance")+theme(plot.title = element_text(hjust = 0.5))

```
